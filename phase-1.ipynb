## NumPy array and basic

```python
import numpy as np
import time
```

```python
arr_1d = np.array([1,2,3,4,5])
print("1D arr: ", arr_1d)

arr_2d = np.array([[1,2,3],[4,5,6]])
print("1D arr: ", arr_2d)
```

### List vs numpy array

```python
py_list = [1,2,3]
print("multiply by 2", py_list*2)

np_list = np.array([1,2,3])
print("multiply by 2", np_list*2)
```

```python
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time:", time.time()- start)
```

```python
start = time.time()
np_array = np.arange(1000000) * 2
print("\n Numpy operation time:", time.time()- start)
```

### create array from scratch

```python
zero = np.zeros((3,4))
print("zeros : \n", zero)

ones = np.ones((2,2))
print("ones : \n", ones)

full = np.full((2,2),2)
print("ones : \n", full)
```
