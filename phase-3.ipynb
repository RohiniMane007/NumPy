```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data sturcture: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 15000, 18000, 21000, 10000], #rest1
    [2, 12000, 13000, 14000, 16000], #rest2
    [3, 20000, 22000, 24000, 28000], #rest3
    [4, 17000, 19000, 20000, 23000], #rest4
    [5, 13000, 15000, 16000, 18000]  #rest5
])

print("=== Sales anlysis====")
print("Sample data for 1st 3 restaurant:", sales_data[0:3])
# print("Sample data for 1st 3 restaurant:", sales_data[:,1:])
```

```python
# Total sales per years

# print(np.sum(sales_data, axis=0)) #sum of each year
print(np.sum(sales_data[:,1:], axis=0)) #sum of each year

# Minimum sales per year
min_sales = np.min(sales_data[:,1:], axis=1)
print(min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:,1:], axis=0)
print(max_sales)

# Average sales per year
avg_sales = np.mean(sales_data[:,1:], axis=1)
print(avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average Cumulative Sales Over Years")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Sum of vector1:", vector1+ vector2)
print("Multiple of vector1:", vector1* vector2)
print("Dot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
